```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# Basic Graph


```python
x = [0, 1,2,3]
y = [0, 2,4,6]

# Resize the graph 
plt.figure(figsize=(5,2), dpi=180)

# Line1
plt.plot(x, y, label='2x', color='#f34e66', marker='.', markersize=6, markeredgecolor='blue')
plt.title('First Graph', fontdict={'fontname': 'Comic Sans MS', 'fontsize': 15})

# Labels for the graph
plt.xlabel('X axis')
plt.ylabel('Y axis')

# Line 2
x2 = np.arange(0, 4, 0.25)

plt.plot(x2[:6], x2[:6]**2, 'g', label='x^2 Acutal')
# Plot the same graph with projections 
plt.plot(x2[5:], x2[5:]**2, 'g--', label='x^2 Projected')

# Graph ticks
# plt.xticks([1, 2, 3])
# plt.yticks([2, 4, 6])

# Legend of the graph
plt.legend()

# Save the graph
# plt.savefig('first-graph.png', dpi=300)
plt.show()
```

## Bar Chart


```python
labels = ['A', 'B', 'C']
values = [1.5, 4, 2.4]

plt.figure(figsize=(4,2), dpi=150)

bars = plt.bar(labels, values, color="#99cfe3")
# bars[0].set_hatch('/')
# bars[1].set_hatch('o')
# bars[2].set_hatch('*')
patterns = ['/', 'o', '*']
for bar in bars:
  bar.set_hatch(patterns.pop(0))


plt.title('Bar Graph Example')

plt.show()
```

# Real world examples

## Gas Price Comparision

### Line Graph

```python
gas_prices = pd.read_csv('./gas_prices.csv')

plt.figure(figsize=(5, 4), dpi=200)
plt.plot(gas_prices.Year, gas_prices.USA, '.-', label="USA")
plt.plot(gas_prices.Year, gas_prices.Canada, '.-', label="Canada")
plt.plot(gas_prices.Year, gas_prices['South Korea'], '.-', label="South Korea")
plt.plot(gas_prices.Year, gas_prices['UK'], '.-', label="UK")

plt.xlabel('Year')
plt.ylabel('USD')

plt.title('Gas Prices over time (USD)')

plt.xticks(gas_prices.Year[::3].tolist()+[2011])
plt.legend()
plt.show()
```

# FIFA Data

## Histograms

```python
fifa_data = pd.read_csv('./fifa-data.csv')

plt.figure(figsize=(5,3), dpi=200)
fifa_data.head(5)

plt.title('Distribution of FIFA Players Skills - 2018')
bins = [i for i in range(40, 110, 10)]
plt.hist(fifa_data.Overall, bins = bins)
plt.ylabel('Number of Players')
plt.xlabel('Skills Level')

plt.xticks(bins)
plt.show()
```

## Pie Chart


```python
# loc -> Lookup on condition
left = fifa_data.loc[fifa_data['Preferred Foot'] == 'Left'].count()[0]
right = fifa_data.loc[fifa_data['Preferred Foot'] == 'Right'].count()[0]

plt.figure(figsize=(3,3), dpi=200)
plt.title('Foot Preference')
labels = ['Left Foot', 'Right Foot']
plt.pie([left, right], labels=labels, autopct='%.2f %%')

plt.show()
```

## Pie chart by weight

```python
# Strip lbs from the string Weight and can convert to KGs if needed
fifa_data.Weight = [int(x.strip('lbs')) if type(x) == str else x for x in fifa_data.Weight]

plt.style.use('ggplot')

plt.figure(figsize=(3,3), dpi=150)

light = fifa_data.loc[(fifa_data.Weight < 125)].count()[0]
light_medium = fifa_data.loc[(fifa_data.Weight >= 125) & (fifa_data.Weight < 150)].count()[0]
medium = fifa_data.loc[(fifa_data.Weight >= 150) & (fifa_data.Weight < 175)].count()[0]
medium_heavy = fifa_data.loc[(fifa_data.Weight >= 175) & (fifa_data.Weight < 200)].count()[0]
heavy = fifa_data.loc[(fifa_data.Weight >= 200)].count()[0]

weights = [light, light_medium, medium, medium_heavy, heavy]
labels = ['< 125', '125-150', '150-175', '175-200', '> 200']
explode = [0.4, 0.2, 0, 0, 0.4]

plt.title('Weight Distribution of FIFA players')

plt.pie(weights, labels=labels, autopct='%.2f %%', pctdistance=0.75, explode=explode)

plt.show()
```
